```python
import os

folder_path = "test"
all_items = os.listdir(folder_path)

audio_paths = [f for f in all_items if os.path.isfile(os.path.join(folder_path, f))]

print(audio_paths)

import whisper
from tqdm import tqdm

model = whisper.load_model("base")
answers = []
for audio_path in tqdm(audio_paths):
    result = model.transcribe(f"{folder_path}/{audio_path}", language="ru")
    answers.append(result["text"])
    # print("Транскрибированный текст:")
    # print(result["text"])
```
